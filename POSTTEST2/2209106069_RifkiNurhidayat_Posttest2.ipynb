# 5. Visualisasikan dataset dengan salah satu grafik. Pilih grafik yang paling cocok untuk memvisualisasikan data tersebut.

Scatter Plot. Kenapa grafik ini paling cocok? berdasarkan data yang ada pada dataset tersebut apabila peningkatan GDP per Capita suatu negara meningkat maka akan membuat kesejahteraan masyarakat meningkat yang berpengaruh besar menambah tingkat kebahagiaan mereka 

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Log GDP per capita', y='Life Ladder', data=df)
plt.title('GDP per Capita vs Happiness Score')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Happiness Score (Life Ladder)')
plt.show()
```

# Tampilkan korelasi dataset menggunakan heatmap.

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap Korelasi Antar Variabel')
plt.show()
```
